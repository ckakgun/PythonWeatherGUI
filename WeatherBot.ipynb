```python
import tkinter as tk
from tkinter import messagebox
import re
import requests

API_KEY = 'YOUR_API_KEY'

def on_enter(event):
    get_weather()

def get_weather():
    city = city_entry.get()
    url = f'http://api.weatherapi.com/v1/current.json?key={API_KEY}&q={city}&aqi=no'
       
    
    if not city or city == "Click here to ask" or not re.match("^[A-Za-z\\s]+$", city):
        messagebox.showwarning("Invalid Input", "Please enter a valid city name.")
        return
    
    response = requests.get(url)
    data = response.json()

    if response.status_code == 200:
        weather = data['current']
        temp = weather['temp_c']
        description = weather['condition']['text']
        location = data['location']['name']
        
        if location.lower() == city.lower():
          
            result = (f"City: {city.capitalize()}\n"
                      f"Temp: {temp}Â°C\n"
                      f"Description: {description.capitalize()}")
        else:
            result = f'Couldn\'t reach the current weather data'
    else:
        error_message = data.get('error', {}).get('message', 'Unknown error')
        result = f'Couldn\'t reach the current weather data: {error_message}'

    chat_area.insert(tk.END, result + "\n")
    city_entry.delete(0, tk.END)

def clear_chat():
    chat_area.delete("1.0", tk.END)
    city_entry.delete(0, tk.END)

root = tk.Tk()
root.title("WeatherBot")
root.configure(bg='black')

chat_area = tk.Text(root, height=30, width=100, bg='white', fg='black')
chat_area.pack(pady = 5)
chat_area.configure(font=('Arial', 12))

city_entry = tk.Entry(root, bg='white', fg='gray', justify= 'center')
city_entry.pack()
city_entry.insert(0, "")  

def on_entry_click(event):
    if city_entry.get() == "Click here to ask":
        city_entry.delete(0, tk.END) 

city_entry.bind("<FocusIn>", on_entry_click)

city_entry.bind("<Return>", on_enter)

get_weather_button = tk.Button(root, text="Get Weather Report", command=get_weather, bg='white', fg='black')
get_weather_button.pack(pady = 5)

clear_button = tk.Button(root, text="Clear", command=clear_chat, bg='white', fg='black')
clear_button.pack(pady=(5, 10))

root.mainloop()
```
